```python
train_df = pd.read_csv('./n_train.csv')
test_df = pd.read_csv('./n_test.csv')
```

```python
train_df = train_df.sort_values(by = ['year', 'climateregions__climateregion', 'loc_group', 'season', 'month']).reset_index(drop=True)
```

```python
test_df = test_df.sort_values(by = ['year', 'climateregions__climateregion', 'loc_group', 'season', 'month']).reset_index(drop=True)
```

```python
def groupby_transform_mean(groupby_cols, source_col):
    train_df['average-{}-by-{}'.format(source_col, '-'.join(groupby_cols))] = train_df.groupby(groupby_cols)[source_col].transform('mean')
    test_df['average-{}-by-{}'.format(source_col, '-'.join(groupby_cols))] = test_df.groupby(groupby_cols)[source_col].transform('mean')    
    
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-wind-h100-14d__wind-hgt-100')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-pevpr-sfc-gauss-14d__pevpr')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-prwtr-eatm-14d__prwtr')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-slp-14d__slp')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-pres-sfc-gauss-14d__pres')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-wind-h850-14d__wind-hgt-850')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-wind-vwnd-925-14d__wind-vwnd-925')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-rhum-sig995-14d__rhum')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-precip-14d__precip')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-wind-uwnd-250-14d__wind-uwnd-250')
groupby_transform_mean(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'], 'contest-wind-vwnd-250-14d__wind-vwnd-250')
```

```python
train_df['average-total-water-mixing-ration'] = (9.81 * train_df['contest-prwtr-eatm-14d__prwtr'] / (train_df['contest-slp-14d__slp'] - train_df['contest-pres-sfc-gauss-14d__pres']))
test_df['average-total-water-mixing-ration'] = (9.81 * test_df['contest-prwtr-eatm-14d__prwtr'] / (test_df['contest-slp-14d__slp'] - test_df['contest-pres-sfc-gauss-14d__pres']))
```

```python
cols = ['contest-wind-h100-14d__wind-hgt-100',
'contest-pevpr-sfc-gauss-14d__pevpr',
'contest-prwtr-eatm-14d__prwtr',
'contest-slp-14d__slp',
'contest-pres-sfc-gauss-14d__pres',
'contest-wind-h850-14d__wind-hgt-850',
'contest-wind-vwnd-925-14d__wind-vwnd-925',
'contest-rhum-sig995-14d__rhum',
'contest-precip-14d__precip',
'contest-wind-uwnd-250-14d__wind-uwnd-250',
'contest-wind-vwnd-250-14d__wind-vwnd-250']
```

```python
for lag_day in [1, 2, 3]:
    for c in cols:
        train_df[f'{c}_lag{lag_day}'] = train_df.groupby(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'])[c].shift(lag_day)
        test_df[f'{c}_lag{lag_day}'] = test_df.groupby(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'])[c].shift(lag_day)
```

```python
for lag_day in [1, 2]:
    for c in cols:
        train_df[f'{c}_roll_avg{lag_day}'] = train_df.groupby(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'])[c].shift(lag_day).rolling(3).mean()
        test_df[f'{c}_roll_avg{lag_day}'] = test_df.groupby(['year', 'climateregions__climateregion', 'loc_group', 'season', 'month'])[c].shift(lag_day).rolling(3).mean()
```

```python
for c in cols:
    train_df[f'delta_{c}_2_1'] = np.where(train_df[f'{c}_lag2'].notnull() & train_df[f'{c}_lag1'].notnull(), abs(train_df[f'{c}_lag1']-train_df[f'{c}_lag2'] ) / train_df[f'{c}_lag2'], np.nan)
    train_df[f'{c}_from_delta2_1'] = np.where(train_df[f'{c}_lag1'].notnull(), train_df[f'{c}_lag1'] + train_df[f'{c}_lag1'] * train_df[f'delta_{c}_2_1'], np.nan)
    
    test_df[f'delta_{c}_2_1'] = np.where(test_df[f'{c}_lag2'].notnull() & test_df[f'{c}_lag1'].notnull(), abs(test_df[f'{c}_lag1']-test_df[f'{c}_lag2'] ) / test_df[f'{c}_lag2'], np.nan)
    test_df[f'{c}_from_delta2_1'] = np.where(test_df[f'{c}_lag1'].notnull(), test_df[f'{c}_lag1'] + test_df[f'{c}_lag1'] * test_df[f'delta_{c}_2_1'], np.nan)
```

```python
feature_names = X_train.columns
selected_cols = []
for score, name in sorted(zip(feature_importances, feature_names), reverse=True):
    if score >= 0.06:
        selected_cols.append(name)
    print('{}: {}'.format(name, score))
```

```python
X = train_df[features]
y = train_df[target[0]]

skf = KFold(n_splits=5, random_state=2048, shuffle=True)
skf.get_n_splits(X, y)

categorical_features_indices = np.where(train_df[selected_cols].dtypes != float)[0]
categorical_feautures = [train_df[features].columns.tolist()[i] for i in categorical_features_indices]

train_index_list = []
test_index_list = []
train_x_list = []
val_x_list = []
train_y_list = []
val_y_list = []

for i, (train_index, test_index) in tqdm(enumerate(skf.split(X, y))):
    train_index_list.append(train_index)
    test_index_list.append(test_index)

    train_x_list.append(X.iloc[train_index])
    val_x_list.append(X.iloc[test_index])
    train_y_list.append(y[train_index])
    val_y_list.append(y[test_index])    
    
def train_catboost(x, y):
    print(x.shape, y.shape)
    
    cat_models = []
    for i in tqdm(range(5)):
        train_x, val_x = train_x_list[i], val_x_list[i]
        train_y, val_y = train_y_list[i], val_y_list[i]
        clf = CatBoostRegressor(iterations=4900, verbose=200, cat_features=categorical_features_indices)
        clf.fit(train_x.values, train_y, eval_set=[(val_x.values, val_y)], verbose=True, early_stopping_rounds=100)
        cat_models.append(clf)
    
    return cat_models
cat_models = train_catboost(X, y)
```

```python
preds = []
for cat_model in tqdm(cat_models):
    preds.append(cat_model.predict(test_df[features]))
s = 0
for pred in preds:
    s += pred
s = s/5
df = pd.read_csv('./sample_solution.csv')
df['contest-tmp2m-14d__tmp2m'] = s
df.to_csv('submission_corr_important.csv', index=False)

best_df = pd.read_csv('best_sub.csv')
y_best = best_df[target[0]].values

from numpy import dot
from numpy.linalg import norm

cos_sim = dot(y_best, s)/(norm(y_best)*norm(s))
print("cos_sim with best submission:", cos_sim)
```
